```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../data/titanic.csv')
data
```

```python
names = []
surnames = []

for name in data['Name']:
    surname, name = name.split(',')
    names.append(name)
    surnames.append(surname)

names = pd.Series(names, name='Name')
surnames = pd.Series(surnames, name='Surname')
```

```python
df = data[data.columns[~data.columns.isin(['Name','PassengerId'])]]
df = pd.concat([df, names, surnames], axis=1)

df
```

```python
# df[df.isna().any(axis=1]
x = df[~df['Cabin'].isna()]['Cabin'].values

(uniq, freq) = np.unique(x, return_counts=True)
(uniq, freq)
```
